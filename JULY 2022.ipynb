```python
def next_position(X, Y, tile_type, step=1.0):
    direction = np.random.randint(1,4)
    delX, delY = 0.0, 0.0
    if (direction==1):
        delY = step
    elif (direction==2):
        delX, delY = 0.5*step*math.sqrt(3), -0.5*step
    else:
        delX, delY = -0.5*step*math.sqrt(3), -0.5*step
    if (tile_type):
        delY *= -1
    return X+delX, Y+delY
```
